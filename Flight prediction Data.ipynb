# Feature Engineering

```python
# now we split the dateof journey to date,month,year wise
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Date']
```

```python
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
# now we split the date of journey .but those ate in string format.
# convert them to int 
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
# now we drop the date_of_journey feature
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# now split the arrivaltime 
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(2)
```

```python
#now we devide the arrival time to hour and minites
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
# now we convert arrival_hour and min into int format
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
# now we drop the arrivaltime feature
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# now split the departure time
#now we devide the arrival time to hour and minites
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
# now we convert arrival_hour and min into int format
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
```

```python
# now we drop the arrivaltime feature
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
# now we convert total stops
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'].value_counts()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
# we map the non stop to 0 and so onnnn
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].value_counts()
```

```python
# now drop the route column from df
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Additional_Info'].unique()
```

```python
# now we split the Duration time
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['duration_hour']
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
# here one record showing 5m  
final_df[final_df['duration_hour']=='5m']
```

```python
# we can drop the records
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
# now we convert to int
final_df['duration_hour']=final_df['duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
# to get duration_hour in minutes
```

```python
final_df['duration_hour']=final_df['duration_hour']*60
final_df.head()
```

```python
# now we split the Duration minutes also
final_df['duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
final_df['duration_min']
```

```python
final_df['duration_min'].head()
```

```python
# so we have 1283 nan values.that means duration in hours only.
# so we can make nan values to 0
```

```python
final_df['duration_min'].fillna(0, inplace=True)
```

```python
final_df['duration_min'].head()
```

```python
final_df.head()
```

```python
# now we convert duration min to int
final_df['duration_min']=final_df['duration_min'].astype(int)
```

```python
# now we combine both duration in hour and min to total_duration minutes
final_df['total_duration_min']=final_df['duration_hour']+final_df['duration_min']
final_df['total_duration_min'].head()
```

```python
# now we will drop the duration hour and min features
final_df.drop(['duration_hour','duration_min'],axis=1,inplace=True)
final_df.head()
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# now we do label encoding for the features
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# apply label encoding to the features
final_df['Airline']=le.fit_transform(final_df['Airline'])
final_df['Source']=le.fit_transform(final_df['Source'])
final_df['Destination']=le.fit_transform(final_df['Destination'])
final_df['Additional_Info']=le.fit_transform(final_df['Additional_Info'])
```
